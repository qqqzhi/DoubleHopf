```python
""" Weijun's Formula """
idx = np.where( (abs(lam.real) < 1e-7) & (lam.imag != 0) )[0]
Q = np.c_[v[:,idx[0]].imag, v[:,idx[0]].real]
i=0
while i < len(lam):
    if i in idx:
        i+=1
        continue
    if lam[i].imag == 0:
        Q=np.c_[Q,v[:,i]]
        i += 1
    else:
        Q=np.c_[Q,v[:,i].imag,v[:,i].real]
        i += 2
            
Q = Q/(abs(np.linalg.det(Q))**(1/n))
P = np.linalg.inv(Q)
J_cs = np.linalg.multi_dot([P,J,Q])
display(pd.DataFrame(J_cs).applymap(lambda x: '{:,.8f}'.format(x)))
```
